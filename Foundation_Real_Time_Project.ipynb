```python
## New Dataset

date=pd.read_csv('drive/MyDrive/Colab Notebooks/Stats and ML datasets/orders.csv')
```

```python
date.head()
```

```python
date.info()
```

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
## extract year

date['date_year']=date['date'].dt.year
date.head()
```

```python
## extract month

date['month']=date['date'].dt.month
date.head()
```

```python
## extract month name

date['month_name']=date['date'].dt.month_name()
date.head()
```

```python
## extract day

date['date_day']=date['date'].dt.day
date.head()
```

```python
## extract day of week

date['date_dow']=date['date'].dt.dayofweek
date.head()         ##0-Mon,1-Tue,2-Wed,3-Thu,4-Fri,5-Sat,6-Sun
```

```python
## extract day of week name

date['date_dow_name']=date['date'].dt.day_name()
date.head()
```

```python
## is is weekend or weekday ?
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),'yes','no')
date.head()
```

```python
## Week

date['week_no']=date['date'].dt.week
date.head()
```

```python
## Quarter

date['Quarter']=date['date'].dt.quarter
date.head()
```

```python
## Semester

date['Semester']=np.where(date['Quarter'].isin([1,2]),1,2)
date.head()
```
